```python
# 1 : while문을 사용해 1부터 1000까지의 자연수 중 3의 배수의 합을 구해 보자.

a = 0
b = 0
while (a<1001): #조건식
    a+=1 #증감식
    if(a%3 == 0):
        b = b + a
        print(a)
print(b)
```

```python
# 6 : 가위바위보 게임에서 승리 횟수가 5회 이상일때 게임이 끝나도록 만들어 보자
# 출력결과 : '승리 : 0/5', 패배 : 1, 비김 : 0'
#'승리 : 1/5', 패배 : 1, 비김 : 0'
#'승리 : 2/5', 패배 : 1, 비김 : 0'
#'승리 : 3/5', 패배 : 1, 비김 : 0'
#'승리 : 3/5', 패배 : 2, 비김 : 0'
#'승리 : 4/5', 패배 : 2, 비김 : 0'
#'승리 : 5/5', 패배 : 2, 비김 : 0'
#'다섯번 이겼습니다. 축하합니다.'

import random
data = ["가위","바위","보"]

print(data.index("가위"))
print(data.index("바위"))
print(data.index("보"))

win = 0 # 초기값
lose = 0 # 초기값
draw = 0 # 초기값
while (True): #조건식

    randomValue = random.choice(data)
    inputData = input("가위 바위 보>>")

    if( ( data.index(randomValue) - data.index(inputData) ) % 3 == 2  ):
        win+=1 #증감식
        

    elif( ( data.index(randomValue) - data.index(inputData) ) % 3 == 1  ):
        lose+=1
        
    elif( ( data.index(randomValue) - data.index(inputData) ) % 3 == 0  ):
        draw+=1

    print(randomValue, "vs", inputData,  ( ( data.index(randomValue) - data.index(inputData) ) % 3 )    )

    print( data.index(randomValue) , data.index(inputData) , ( ( data.index(randomValue) - data.index(inputData) ) % 3 ) )


    print('승리 : ', win , '/5, 패배 : ' , lose , ', 비김 : ' , draw)
        
    if(win == 5):
        print('다섯번 이김')
        break # 반복문 강제 종료


    '''
    0가위 ok
        1-0-> 1
        2-0-> 2

    1바위 
        2-1 -> 1
        0-2 -> -2 2

    2보
        0-2 -> 

    '''
```
